```python
# drop datetime column since we have the splitted date and time columns
df.drop(['Incident Datetime'], axis =1, inplace=True)
df.drop(['Incident Year'], axis =1, inplace=True)
```

```python
# drop CAD Number column
# Linked to Incident Number + reports not filed thru the DEM have no number
df.drop(['CAD Number'], axis =1, inplace=True)
```
